```python
from pyforest import *
from sklearn.model_selection import KFold
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data = data.drop(labels='Unnamed: 32',axis=1)
data['diagnosis'] = pd.factorize(data['diagnosis'])[0]
```

```python
data_arr = np.array(data)
```

```python
data
```

```python
X = data_arr[:,2:]
Y = data_arr[:,1]
```

```python
kf = KFold(n_splits=5,shuffle=True)
kf.get_n_splits(X)
```

```python
model = GaussianNB()
```

```python
acc1 = 0
for train_in, test_in in kf.split(X):
    X_train, X_test = X[train_in], X[test_in]
    Y_train, Y_test = Y[train_in], Y[test_in]
    
    model.fit(X_train,Y_train)
    y_pred = model.predict(X_test)
    acc1 += accuracy_score(Y_test,y_pred)
print("Average Accuracy :", acc1/5)
```

```python
model = LogisticRegression()
```

```python
acc2 = 0
for train_in, test_in in kf.split(X):
    X_train, X_test = X[train_in], X[test_in]
    Y_train, Y_test = Y[train_in], Y[test_in]
    
    model.fit(X_train,Y_train)
    y_pred = model.predict(X_test)
    acc2 += accuracy_score(Y_test,y_pred)
print("Average Accuracy :", acc2/5)
```

```python
plt.hist([acc1/5,acc2/5])
```
